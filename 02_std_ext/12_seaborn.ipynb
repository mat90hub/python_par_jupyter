# L'extension [Seaborn](https://seaborn.pydata.org/index.html)

L'extension [Seaborn](https://seaborn.pydata.org/index.html)se base sur l'extension historique [matplotlib.pyplot](https://matplotlib.org/stable/api/pyplot_summary.html) pour se focaliser sur le visualisation de données statistiques. Elle offre aussi des thèmes par défaut aux couleurs pastelles, qui sont très agréables à regarder et s'intègrent bien pour un rapport de qualité. Elle est aussi construite pour s'intégrer très bien avec des structures de données traitées par [pandas](https://pandas.pydata.org/docs/user_guide/index.html#user-guide).

Son intérêt va résider dans l'obtention simplifier des différents types de diagrammes utilisés en statistique. Si on veut ensuite personnaliser plus loin le résultat obtenu, il faudra sûrement faire appel aux fonctionalités de matplotlib, qui existent toujours puisque Seaborn est une extension de matplotlib.

# Les affichages possibles d'une distribution simple.

Nous construisons une liste de cent valeur prise selon une distribution normale centrée.


```python
import numpy as np

rng = np.random.default_rng(10) # création du générateur de nombres aléatoires
x = rng.standard_normal(1000)   # création de l'échantillon
x[0:5]
```

Les représentations statistiques de cet échantillon peuvent être :
- un histogramme montrant les effectifs par classe ;
- la même chose sous forme de courbe ;
- ou un histogramme de effectif cumulé.

on utilisera pour cela les différentes façon (*kind*) de la fonction [displot](https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot)

```python
import seaborn as sns
sns.set_theme(context='notebook', style='darkgrid')
```

```python
sns.displot(x, kind="hist") # 4 types: kde, hist, ecdf
```

```python
sns.displot(x, kind='kde')
```

```python
sns.displot(x, kind='ecdf')
```

[L'API](https://seaborn.pydata.org/api.html) de seaborn donne d'un coup d'oeil les différentes possibilités.
