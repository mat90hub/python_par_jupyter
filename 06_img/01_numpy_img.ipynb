L'extension permettant de visualiser des images s'appelle [PIL](https://he-arc.github.io/livre-python/pillow/index.html) pour `Python Imaging Library`. Cette extension est maintenant remplace par [Pillow](https://pillow.readthedocs.io/en/stable/) qui un *fork* qui remplace *PIL* mais avec les mmes commandes. On ouvre une image de la faon suivante et on peut lire les principales caractristiques.

```python
from PIL import Image
image = Image.open('raton.png')

print(image.format)
print(image.size)
print(image.mode)
```

L'image s'affiche par exemple avec [matplotlib.pyplot.imshow](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html).

```python
import matplotlib.pyplot as plt
plt.axis('off')
plt.imshow(image)
plt.show()
```
